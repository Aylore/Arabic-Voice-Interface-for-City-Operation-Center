```python
import pandas as pd
import string
import string
import unicodedata as ud
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_colwidth', 500)
```

```python
df = pd.read_csv("csv files/azure_en_results.csv",index_col=0)
```

```python
df.head()
```

```python
# Function to remove punctuation and normalize letters
def preprocess_text(text):
    # Remove punctuation
    text = text.translate(str.maketrans('', '', string.punctuation))
    
    text = ''.join(c for c in text if not ud.category(c).startswith('P'))
    
    text = text.lower()
    # Tokenize the text
    return text
```

```python
# Apply preprocessing to predicted text and true text columns
df['pred'] = df['pred'].apply(preprocess_text)
df['true label'] = df['true label'].apply(preprocess_text)

# Print the processed DataFrame
df.head(50)
```

# Utility Functions

```python
import cv2
import moviepy.editor as mp

def crop_mp4(video_path, start: int, end: int, output_path=None):
    """
    Crop a video from the specified start time to end time,
    and save it as a new video file with audio included.
    """
    # Read the input video
    input_path = video_path
    video = mp.VideoFileClip(input_path)

    # Set the start and end time for the desired subset (in seconds)
    start_time = start
    end_time = end

    # Crop the video
    cropped_video = video.subclip(start_time, end_time)

    # Set the output path
    output_path = video_path[:-4] + '_cropped.mp4' if not output_path else output_path

    # Write the cropped video with audio to the output file
    cropped_video.write_videofile(output_path, codec="libx264", audio=True)

    # Close the video file
    video.reader.close()

    # Close the audio file
    video.audio.reader.close_proc()

    return output_path
```

```python

def get_video_length(video_path):
    # Open the video file
    video = cv2.VideoCapture(video_path)

    # Get the total number of frames in the video
    total_frames = int(video.get(cv2.CAP_PROP_FRAME_COUNT))

    # Get the frames per second (FPS) of the video
    fps = video.get(cv2.CAP_PROP_FPS)

    # Calculate the duration of the video in seconds
    duration = total_frames / fps

    # Convert duration to minutes and seconds
    minutes = int(duration / 60)
    seconds = int(duration % 60)

    # Close the video file
    video.release()

    return minutes*60  + seconds
```
